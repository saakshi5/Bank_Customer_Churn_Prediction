<font size="6">7.Splitting Data into x and y<font>

```python
# splitting x and y into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

<font size='6'>8.Importing Required Libraries for Model</font>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
```

```python
# creating instance of class
lg=LogisticRegression()
knn=KNeighborsClassifier()
svm=SVC()
dt=DecisionTreeClassifier()
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    print('Classification Report')
    print(classification_report(ytest,ypred))
    print('-'*50)
    print('Training accuracy',model.score(xtrain,ytrain))
    print('Testing accuracy',model.score(xtest,ytest))
    print('accuracy score',accuracy_score(ypred,ytest))
```

Building Logistic Regression Model

```python
mymodel(lg)
```

From the classification report of the logistic regression model we get following insights:
The model performs well in predicting non-churned cases (class 0) with high precision (80%), recall (98%), and F1-score (88%). This suggests that when the model predicts a customer will not churn, it is correct most of the time and captures almost all actual non-churned cases.
However, the model struggles in predicting churned cases (class 1) with low precision (42%), recall (6%), and F1-score (11%). This indicates that when the model predicts a customer will churn, it is often incorrect (low precision), and it misses many actual churned cases (low recall).
The overall accuracy of the model is 0.79.
In this case, the model's testing accuracy (0.79) is comparable to the training accuracy (0.79), suggesting the model is not significantly overfitting.

```python
#Predicting probabilities 
ypredprob=lg.predict_proba(xtest)[:,1]
```

```python
ypredprob
```

```python
#Drawing roc_auc curve
fpr,tpr,thres=roc_curve(ytest,ypredprob)
plt.plot(fpr,tpr)
plt.show()
```

Thus, the poor performance of the logistic regression model, particularly in predicting churned cases (class 1), will likely result in a less optimal ROC curve.

<font size='6'>9.Model Tuning</font>

KNN Model Tuning

```python
#calculating best k values
ac_lst=[]
for i in range(1,21):
    kn=KNeighborsClassifier(n_neighbors=i)
    kn.fit(xtrain,ytrain)
    ypredd=kn.predict(xtest)
    ac=accuracy_score(ytest,ypredd)
    ac_lst.append(ac)
```

```python
#identifying best value of k by plotting
plt.plot(range(1,21),ac_lst,marker='o')
```

Here, best value of k is 12

```python
#using k=12
knn.fit(xtest,ytest)
ypredd=knn.predict(xtest)
print('Classification Report')
print(classification_report(ytest,ypredd))
print('-'*50)
print('Training accuracy',knn.score(xtrain,ytrain))
print('Testing accuracy',knn.score(xtest,ytest))
print('accuracy score',accuracy_score(ypredd,ytest))
```

The updated KNN model with tuned hyperparameters (n_neighbors = 12) shows significant improvements in predicting churned cases (class 1) compared to the previous model.
The selected value of k strikes a good balance between bias and variance, leading to improved generalization performance on unseen data.
The training accuracy (0.75) is slightly lower than the testing accuracy (0.81), indicating that the model performs well on unseen data and does not overfit.
The accuracy also incresed from 76% to 82%.
